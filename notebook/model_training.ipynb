# Tech used:
Will be using a TFIDF vectorizer and Passive Aggressive classifier from Sklearn 

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import PassiveAggressiveClassifier
from sklearn.metrics import confusion_matrix, accuracy_score
```

```python
df = pd.read_csv(r"/root/Cyber/FakeNewsDetection/datasets/news.csv")
```

```python
df.head()
```

```python
df.shape
```

```python
df.columns=df.columns.str.strip()
```

```python
X=df["text"]
Y=df["label"]
```

```python
X,Y
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,Y,test_size=0.2,random_state=42)
```

```python
tfid_vector= TfidfVectorizer(stop_words="english", max_df=0.7)
tfidf_train= tfid_vector.fit_transform(X_train)
tfidf_test=tfid_vector.transform(X_test)
```

```python
#Setting up the passive Aggressive Classifier
pac = PassiveAggressiveClassifier(max_iter=50)
pac.fit(tfidf_train,y_train)


y_pred=pac.predict(tfidf_test)
```

```python
accuracy= accuracy_score(y_test,y_pred)
print(f"Accuracy: {accuracy}")
```

```python
pac.get_params()
```

```python
confusion_matrix(y_test,y_pred, labels=["FAKE","REAL"])
```

```python
#Save the model 
import joblib

# Save TF-IDF vectorizer
joblib.dump(tfid_vector, "tfidf_vector.pkl")

# Save trained model
joblib.dump(pac, "fake_news_model.pkl")
```
